```python
def walk(steps):
    for step in range(1,steps+1):
        print(f"You take step number {step}")
```

```python
walk(100)
```

## recursively

```python
def walk(steps):
    print(f"You take step number {steps}")
    walk(steps-1)
    
```

```python
walk(100)
```

```python
def walk(steps):
    if steps == 0:
        return
    print(f"You take step number {steps}")
    walk(steps-1)
```

```python
walk(100)
```

```python
def walk(steps):
    if steps == 0:
        return
    walk(steps-1)
    print(f"You take step number {steps}")
```

```python
walk(100)
```

# find the factorial 

iteratively

```python
def factorial(x):
    result = 1
    
    if x>0:
        for y in range (1, x+1):
            result *= y
        return result

factorial(10)
```

recursively

```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return x*factorial(x-1)

factorial(10)
```
