```python
from tkinter import *

class Counter:

    def __init__(self, master):
        self.master = master
        master.title("Counter")
        master.geometry("300x150")

        self.count = 0
        # Corrected typo here
        self.label = Label(master, text=self.count)
        self.label.pack()

        # Label will be updated when buttons are clicked
        self.incr_button = Button(master, text="Tambah", command=self.increase)
        self.incr_button.pack()
        self.decr_button = Button(master, text="Kurang", command=self.decrease)
        self.decr_button.pack()

    # refer ke method command td ke sini
    # kenapa pake self.method, ga method()? karena ini tu kg manggil func
    # /methodnya, tp reference
    def increase(self):
        self.count += 1
        self.label['text'] = self.count

    def decrease(self):
        self.count -= 1
        self.label['text'] = self.count

root = Tk()
counter = Counter(root)
root.mainloop()
```
